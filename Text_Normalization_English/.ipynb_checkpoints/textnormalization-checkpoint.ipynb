```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading the train data using pandas read csv function
df_train =pd.read_csv("en_train.csv")
df_train.head()
```

```python
df_train.shape
```

```python
train_token_per_sent = df_train.groupby('sentence_id')['sentence_id'].count()
train_token_per_sent.head()
```

```python
train_token_per_sent.shape
```

```python
train_token_per_sent.describe()
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,10))
count_length_fig = sns.countplot(train_token_per_sent, ax=ax)
for item in count_length_fig.get_xticklabels():
    item.set_rotation(90)
```

```python
df_test = pd.read_csv("en_test.csv")
df_test.head()
```

```python
test_token_per_sent = df_test.groupby('sentence_id')['sentence_id'].count()
test_token_per_sent.head()
```

```python
test_token_per_sent.describe()
```

```python
fig,ax = plt.subplots(1,1,figsize=(20,10))
count_length_fig = sns.countplot(test_token_per_sent, ax=ax)
for item in count_length_fig.get_xticklabels():
    item.set_rotation(90)
```

```python
#Longest train sentence 
id_max = train_token_per_sent[train_token_per_sent == train_token_per_sent.max()].index.values
id_max
```

```python
long_train = df_train[df_train['sentence_id']==id_max[0]].before.values.tolist()
long_train =''.join(long_train)
long_train
```

```python
id_min = train_token_per_sent[train_token_per_sent == train_token_per_sent.min()].index.values
short_train = df_train[df_train['sentence_id']==id_min[0]].before.values
short_train = ''.join(short_train)
short_train
```

```python
#Coming to tokens 
unique_class_train = df_train.groupby('class')['class'].count()
print unique_class_train
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,10))
sns.barplot(x=unique_class_train.values,y=unique_class_train.index)
```

```python
#counts of individual classess
df_train.groupby('class')['class'].count()
```

```python
#most before words in electronic class
most_electronic_class=df_train[df_train['class']=='ELECTRONIC'].groupby("before")['before'].count().sort_values(ascending=False).head(10)
fig,ax = plt.subplots(1,1,figsize=(10,10))
sns.barplot(x=most_electronic_class.index,y=most_electronic_class.values)
```
